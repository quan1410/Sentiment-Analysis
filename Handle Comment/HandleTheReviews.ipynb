```python
import nltk # using textBlob and tensoflow (CNN and LSTM )
import pprint
import tensorflow as tf
import pandas as pd
nltk.download('punkt')
from nltk.tokenize import word_tokenize, sent_tokenize
# some source for this dic 
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
data_train = pd.read_csv('C:/Project/movie_reviews/movie_reviews.csv')
data_test = pd.read_csv('C:/Project/test_data.csv')
```

```python
print(data_train.shape) # trainning data to feed module 
print(data_test.shape) # to predict data 
```

```python
stopword = set('for in and to a an of the'.split(' '))
stopword
```

### Prepare for trainning data


```python
import numpy as np 
input_train= data_train['text']
input_test = data_test['text']
label_train = data_train.index.values
label_test = data_test.index.values
    
    
```

```python
data_train
```

```python
# Build my dictionary
vocab_size = 10000
embedding = 64
max_length = 140 


tokenizer = Tokenizer(num_words= vocab_size, oov_token= "<OOV>")
tokenizer.fit_on_texts(input_train)
tokenizer.word_index

```

```python
input_test
```

```python

input_train = tokenizer.texts_to_sequences(input_train)
padd_train = pad_sequences(input_train, maxlen= max_length , truncating= 'post' , padding= "post")
padd_train
```

```python

input_test = tokenizer.texts_to_sequences(input_test)
padd_test = pad_sequences(input_test, maxlen= max_length , truncating= 'post' , padding= "post")
padd_test
```

### Start to build module 

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Embedding(vocab_size, embedding, input_length = max_length ))
model.add(Flatten()) # converrt to vecto
model.add(Dense(10, activation = 'relu')) # if elements < 0 default it equal 0
model.add(Dense(1, activation = 'sigmoid')) # convert value between 0 and 1 
model.compile(loss = 'binary_crossentropy' , optimizer='adam' , metrics = ['acc'] )
model.summary()
```

```python
print(padd_train.shape)
print(padd_test.shape)
```

```python
# Trainning the module
model.fit(padd_train , label_train ,epochs= 10, validation_data= (padd_test , label_test ))
```
